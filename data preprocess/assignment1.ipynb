```python
# 删除属性
# 因为
def delete_feature(df):
    del_feature, save_feature = [], []
    
    for col_name in df.columns:
        length = len(df[col_name])
        num_invalid_value = list(df[col_name].isna()).count(True)
        loss_rate = num_invalid_value / length
        if loss_rate > 0.15:
            del_feature.append(col_name)
        else:
            save_feature.append(col_name)
            
    return del_feature, df[save_feature]
```

```python
del_feature, dfll = delete_feature(data_df)
print(del_feature)
dfll
```

```python
# 删除样本
# 只删除数据缺失的样本
def delete_sample(df):
    df_ = df.dropna()
    return df_
```
